```python
correlation_matrix = hc.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
age = hc['age']
plt.title('Boxplot of Age by Diabetes Status')
sns.boxplot(data=hc,x='diabetes',y='age')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='hypertension', hue='diabetes', data=hc)
plt.title('Countplot of Hypertension by Diabetes')
plt.subplot(1, 2, 2)
sns.countplot(x='heart_disease', hue='diabetes', data=hc)
plt.title('Countplot of Heart Disease by Diabetes')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,4))
sns.boxplot(x = 'blood_glucose_level', ax = axes[0], data = hc)
sns.boxplot(x = 'bmi', ax = axes[1], data = hc)
```

```python
sns.countplot(data=hc,x='smoking_history',hue='diabetes')
```

```python
fig, axes = plt.subplots(3,2, figsize=(15,12),sharey=True)
sns.countplot(x='gender',hue='diabetes',data=hc,ax=axes[0,0])
sns.countplot(x='heart_disease',hue='diabetes',data=hc,ax=axes[0,1])
sns.countplot(x='hypertension',hue='diabetes',data=hc,ax=axes[1,0])
sns.countplot(x='blood_glucose_level',hue='diabetes',data=hc,ax=axes[2,0])
sns.countplot(x=hc['smoking_history'],ax=axes[2,1])
sns.countplot(x=hc['HbA1c_level'],ax=axes[1,1])
```
